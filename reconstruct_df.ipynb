```python
def reconstruct_dataframe(new_df):
    new_df=new_df.dropna(how='all')
    for index, row in new_df.iterrows():
        if pd.isna(new_df.loc[index,"Name"]):
            new_sub1=new_df.loc[index-1,"Subjects1"]+new_df.loc[index,"Subjects1"]
            new_sub2=new_df.loc[index-1,"Subjects2"]+new_df.loc[index,"Subjects2"]
            new_df.loc[index-1,"Subjects1"]=new_sub1
            new_df.loc[index-1,"Subjects2"]=new_sub2
    new_df=new_df.dropna(subset=["Name"])
    new_df.reset_index()

    return new_df
```
